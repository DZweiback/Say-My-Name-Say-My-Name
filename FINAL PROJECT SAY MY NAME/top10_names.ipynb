```python
import pandas as pd
data = pd.read_csv("BabyNames_States.csv")

data[-600:-1]
```

```python
grouped = data.groupby("Name").count().reset_index()
top10_names = grouped.sort_values("Count",  ascending = False).head(10)
top10_names = grouped.sort_values("Year",  ascending = False).head(10)
```

```python
top10_names = top10_names[["Year", "Name", "Count"]].reset_index(drop = True)
top10_names
```

```python
top10_names.to_csv("top10_names.csv")
```
