# Load and Convert Baby Names and Year of Birth

```python
# load and read name files; add year column based on file name

fileList=glob.glob("Baby Names/*.txt")

#print(fileList)
#def addField(fileList):
for filename in fileList:
    #print(filename)
    df=pd.read_csv(filename, header=None)           
    df['yr']=[filename.rsplit("b",1)[1]]*df.shape[0]
    #print(df['yr'])
    df.to_csv(filename+".csv", index=None, header=None)
```

```python
# Add Column Headings, Concatenate files and merge to single csv file "Concatenated.csv"

outfile="BabyNames.csv"
fileList=glob.glob("Baby Names/*.csv")
dfList=[] # create an empty list
colnames=['Name', 'Gender', 'Count', 'Year']

for filename in fileList:
    print(filename)
    df=pd.read_csv(filename, header=None)
    dfList.append(df)
    
concatDf=pd.concat(dfList, axis=0)
concatDf.columns=colnames
```

```python
# Sample output from the "Concatenated.csv" file
concatDf.head()
```

```python
# CODE for splitting .txt from Year field - Leaving only the year
newYr = concatDf["Year"].str.split(".",n=1, expand=True)
#newYr.head()
concatDf["Year"]=newYr[0]
concatDf["txt"]=newYr[1]
concatDf.drop(columns=['txt'], inplace=True)
concatDf.head()
```

```python
concatDf.tail()
```

```python
# Pass dataframe to a .csv file

concatDf.to_csv(outfile, index=None)
# len(concatDf)
```

```python
# For State files Add Column Headings, Concatenate files and merge to single csv file "Concatenated.csv"

outfile="BabyNames_States.csv"
fileList=glob.glob("Baby Names with States/*.txt")
dfList=[] # create an empty list
colnames=['State','Gender', 'Year', 'Name', 'Count']

for filename in fileList:
    print(filename)
    df=pd.read_csv(filename, header=None)
    dfList.append(df)


StatesDF=pd.concat(dfList, axis=0)
StatesDF.columns=colnames
StatesDF.head()
```

```python
# Insert Credits Data from csv file
credits="Cleaned_tmdb_5000_credits.csv"
creditsDF=pd.read_csv(credits)
creditsDF.drop(columns=['crew'], inplace=True)
creditsDF.head()
```

```python
# Convert 'cast' list of dictionaries in JSON to a dataframe; Loop through all of the dictionaries and append to the dataframe.
# Add the movie_id for each record and Drop the 'credit_id' field

counter = 0 
#print(type(creditsDF))
character_df = pd.DataFrame(columns=['cast_id', 'character', 'gender', 'id', 'name', 'order', 'movie_id', 'title'])
for row in creditsDF.iterrows():     
    counter = counter + 1
    #print(counter)                                 #Keep to show counter for debugging
    castListDF = row[1][2]
    #print(row)                                     #Keep to print each row for debugging
    try:
        ch_data = json.loads(castListDF)
    except Exception as e:
        print(row)
        print('Sorry, information not available')
        pass
    ch_df = pd.DataFrame(ch_data)                   #Builds a new dataframe for each movie "ch_df" 
    ch_df['movie_id'] = row[1]['movie_id']          #Adds the movie id to each row in the "ch_df" dataframe
    ch_df['title'] = row[1]['title']                #Adds the movie title to each row in the dataframe
    ch_df.drop(columns=['credit_id'], inplace=True) #Drops the credit_id from the dataframe
    character_df=character_df.append(ch_df)         #Appends the new dataframe to the main "character_df"   
    #if counter >= 500:                             #Keep to break after short run for debugging
        #break
#character_df.head   
```

```python
# CODE for splitting ['character'] name to first name only
newName = character_df["character"].str.split(" ",n=2, expand=True)
#newName.head()
character_df["First"]=newName[0]
character_df["Second"]=newName[1]
character_df["Third"]=newName[2]
#concatDf.drop(columns=['Second'], inplace=True)
character_df.head()
```

```python
# Export dataframe to "characters.csv" file
character_df.to_csv("characters.csv", index=True, header=True)
```
